```python
import pandas as pd
import numpy as np
```

```python
dataset=pd.read_csv("kidney_disease.csv")
```

```python
dataset
```

```python
dataset.isna().sum()
```

```python
quan=[]
qual=[]
for columnname in dataset:
    if dataset[columnname].dtype=="object":
        qual.append (columnname)
        
    else:
        quan.append (columnname)
        
print("qual=",qual) 
print("quan=",quan)
```

```python
#to replace nan values in quan
from sklearn .impute import SimpleImputer
model=SimpleImputer(missing_values=np.nan,strategy="mean")
model.fit(dataset[quan])
model.transform(dataset[quan])
```

```python
dataset[quan].isna().sum()
```

```python
quan
```

```python
for columnname in quan:
    dataset[columnname].fillna(dataset[columnname].mean(),inplace=True)
```

```python
dataset[quan].isna().sum()
```

```python
pd.DataFrame(dataset[quan])
```

```python
from sklearn .impute import SimpleImputer
model=SimpleImputer(missing_values=np.nan,strategy="most_frequent")
model.fit(dataset[qual])
model.transform(dataset[qual])
```

```python
dataset[qual].isna().sum()
```

```python
qual
```

```python
for columnname in qual:
    dataset[columnname].fillna(dataset[columnname].mode()[0],inplace=True)
```

```python
dataset[qual].isna().sum()
```

```python
dataset.isna().sum()
```

```python
dataset.to_csv("ckd_preprocessed_data.csv")
```
